```python
import os
import numpy as np
import pandas as pd
```

```python
all_files = os.listdir("../Metrics/")

csv_files = [f for f in all_files if f.startswith('metrics_anisot') and f.endswith('.csv')]

dfs = []

for file in csv_files:
    file_path = os.path.join("Metrics/", file)
    df = pd.read_csv(file_path)
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)

merged_df.head()
```

```python
numeric_df = merged_df.select_dtypes(include=[np.number])

means = numeric_df.mean()
stds = numeric_df.std()

summary_df = pd.DataFrame(columns=numeric_df.columns)

for column in numeric_df.columns:
    summary_df.at[0, column] = f"{means[column]:.2f} ± {stds[column]:.2f}"

summary_df.head()
```
